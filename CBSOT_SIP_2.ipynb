```python
import os
import numpy as np

# The embedding dimension is 384, as observed in previous cell outputs.
EXPECTED_EMBEDDING_DIM = 384

embeddings = None
embedding_file_path = 'paper_embeddings.npy'

if os.path.exists(embedding_file_path):
  print('Attempting to load saved embeddings...')
  try:
    loaded_embeddings = np.load(embedding_file_path)
    # Check if the number of rows matches the current DataFrame length
    # and if the embedding dimension is correct.
    if loaded_embeddings.shape[0] == len(df) and loaded_embeddings.shape[1] == EXPECTED_EMBEDDING_DIM:
      embeddings = loaded_embeddings
      print('Embeddings loaded successfully.')
    else:
      print(f'Saved embeddings have an incorrect shape {loaded_embeddings.shape}, expected ({len(df)}, {EXPECTED_EMBEDDING_DIM}).')
      print('Deleting old embeddings file and regenerating...')
      os.remove(embedding_file_path)
  except ValueError as e:
    print(f'Error loading embeddings from {embedding_file_path} due to shape mismatch: {e}.')
    print('Deleting old embeddings file and regenerating...')
    if os.path.exists(embedding_file_path): # Ensure it exists before trying to remove
        os.remove(embedding_file_path)
  except Exception as e:
    print(f'An unexpected error occurred while loading embeddings from {embedding_file_path}: {e}.')
    print('Deleting old embeddings file and regenerating...')
    if os.path.exists(embedding_file_path): # Ensure it exists before trying to remove
        os.remove(embedding_file_path)

if embeddings is None:
  print('Generating embeddings...')
  embeddings = model.encode(df['Paper_text'].to_list(), batch_size=32, show_progress_bar=True)
  np.save(embedding_file_path, embeddings)
  print('Embeddings saved successfully.')
```

```python
import os
import faiss

faiss_index_file_path = 'paper_faiss.index'
index = None

if os.path.exists(faiss_index_file_path):
  print('Attempting to load existing FAISS index...')
  try:
    index = faiss.read_index(faiss_index_file_path)
    print('FAISS index loaded successfully.')
  except RuntimeError as e:
    print(f"Error loading FAISS index from {faiss_index_file_path}: {e}.")
    print('Deleting corrupted FAISS index file and regenerating...')
    os.remove(faiss_index_file_path)
  except Exception as e:
    print(f"An unexpected error occurred while loading FAISS index from {faiss_index_file_path}: {e}.")
    print('Deleting old FAISS index file and regenerating...')
    os.remove(faiss_index_file_path)

if index is None:
  print('Creating new FAISS index...')
  # The 'embeddings' variable from kernel state (0J-iRcwGZjph) is (15000, 384)
  # and should be used for building the index.
  faiss.normalize_L2(embeddings)
  index = faiss.IndexFlatIP(embeddings.shape[1]) # Use the actual embedding dimension
  index.add(embeddings)
  faiss.write_index(index, faiss_index_file_path)
  print('FAISS index saved successfully.')
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
useful_labels = [
    "ORG", "PERSON", "GPE", "LOC",
    "DATE", "PRODUCT", "EVENT",
    "WORK_OF_ART", "NORP", "FAC"
]

def find_entities(text):

    doc = nlp(text)

    entities = []
    seen = []

    for entity in doc.ents:

        if entity.label_ in useful_labels:

            if entity.text not in seen:

                entities.append({
                    "name": entity.text,
                    "type": entity.label_,
                    "description": spacy.explain(entity.label_)
                })

                seen.append(entity.text)

    return entities
```

```python
example_text = "Google was founded by Larry Page and Sergey Brin in California on September 4, 1998. The company is headquartered in Mountain View."
entities = find_entities(example_text)
print("Extracted Entities:")
for entity in entities:
    print(f"\nName: {entity['name']}")
    print(f"Type: {entity['type']}")
    print(f"Description: {entity['description']}")
```

```python




def search_with_keywords_and_entities(query, k=5):

    # Convert query into embedding
    query_vector = model.encode([query])
    faiss.normalize_L2(query_vector)

    scores, indexes = index.search(query_vector, k)
    for score, paper_index in zip(scores[0], indexes[0]):
        paper = df.iloc[paper_index]
        title = paper["title"]
        abstract = paper["abstract"]
        full_text = title + " " + abstract
        keywords = kw_model.extract_keywords(
            full_text,
            keyphrase_ngram_range=(1, 3),
            stop_words="english",
            top_n=10
        )
        entities = find_entities(full_text)
        summary = summarizer(
            abstract,
            max_length=120,
            min_length=40,
            do_sample=False
        )
        print("=" * 80)
        print(f"Similarity Score : {score:.4f}")
        print("\nTitle")
        print(title)
        print("\nAbstract")
        print(abstract)
        print("\nSummary")
        print(summary[0]["summary_text"])
        print("\nKeywords")
        if keywords:
            for keyword, _ in keywords:
                print(f"- {keyword}")
        else:
            print("No keywords found.")
        print("\nNamed Entities")
        if entities:
            for entity in entities:
                print(f"- {entity['name']} ({entity['type']})")
        else:
            print("No named entities found.")
        print("\n")
```

```python
search_with_keywords_and_entities("Deep learning in medical imaging", k=5)
```
